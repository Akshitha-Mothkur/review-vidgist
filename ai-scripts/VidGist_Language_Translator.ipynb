```python
from transformers import pipeline

# Translation models for Indian languages
language_models = {
    "hindi": "Helsinki-NLP/opus-mt-en-hi",
    "tamil": "Helsinki-NLP/opus-mt-en-ta",
    "telugu": "Helsinki-NLP/opus-mt-en-te",
    "bengali": "Helsinki-NLP/opus-mt-en-bn",
    "kannada": "Helsinki-NLP/opus-mt-en-kn",
    "malayalam": "Helsinki-NLP/opus-mt-en-ml",
    "marathi": "Helsinki-NLP/opus-mt-en-mr",
    "gujarati": "Helsinki-NLP/opus-mt-en-gu",
    "punjabi": "Helsinki-NLP/opus-mt-en-pa",
    "odia": "Helsinki-NLP/opus-mt-en-or",
    "urdu": "Helsinki-NLP/opus-mt-en-ur",
    "assamese": "Helsinki-NLP/opus-mt-en-as"
}

# English text
sample_text = """
Artificial Intelligence (AI) refers to the simulation of human intelligence in machines that are programmed to think like humans and mimic their actions.
It can also be applied to any machine that exhibits traits associated with a human mind such as learning and problem-solving.
The ideal characteristic of artificial intelligence is its ability to rationalize and take actions that have the best chance of achieving a specific goal.
Machine learning, a subset of AI, refers to the concept that computer programs can automatically learn from and adapt to new data without being assisted by humans.
Deep learning techniques enable this automatic learning through the absorption of huge amounts of unstructured data such as text, images, or video.
"""

# Show available Indian languages
print("Available Indian languages:")
for idx, lang in enumerate(language_models.keys(), start=1):
    print(f"{idx}. {lang.capitalize()}")

# Choose language
choice = int(input("Enter the number of the language you want to translate to: "))
selected_language = list(language_models.keys())[choice - 1]

# Load translation pipeline
print(f"\n🔄 Translating to {selected_language.capitalize()}...\n")
translator = pipeline("translation", model=language_models[selected_language])

# Translate
translation = translator(sample_text, max_length=512)[0]['translation_text']

# Show output
print(f"--- {selected_language.capitalize()} Translation ---")
print(translation)

# Save to file
file_name = "translation.txt"
with open(file_name, "w", encoding="utf-8") as f:
    f.write(translation)

print(f"\n✅ Translation saved to {file_name}")
```
