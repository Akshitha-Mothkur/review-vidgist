```python
from transformers import pipeline

# Load the Hugging Face summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def format_bullet_points(text):
    return "🔹 " + text.replace('. ', '.\n🔹 ')
```

```python
def generate_3_level_summary(transcript):
    if len(transcript) > 3000:
        transcript = transcript[:3000]

    # 🏷️ Generate content-specific title
    title = summarizer(transcript, max_length=12, min_length=5, do_sample=False)[0]['summary_text']
    title = title.strip().rstrip('.')

    # 🔹 Short Summary
    short_raw = summarizer(transcript, max_length=45, min_length=20, do_sample=False)[0]['summary_text']
    short_bullets = format_bullet_points(short_raw)

    # 🔸 Medium Summary
    medium_raw = summarizer(transcript, max_length=100, min_length=60, do_sample=False)[0]['summary_text']
    medium_section = f"""
🧩 **Concept Breakdown**
{medium_raw}

💡 **Example**
Imagine applying this concept in a scenario like daily recommendations on YouTube, or AI-generated resumes.
"""

    # 🔶 Detailed Summary
    long_raw = summarizer(transcript, max_length=200, min_length=100, do_sample=False)[0]['summary_text']
    detailed_section = f"""
📘 **Detailed Overview**
{long_raw}

🧠 **What This Means**
This concept plays a crucial role in broader contexts like automation, education, or responsible tech development.

🔚 **Conclusion**
To conclude, embracing this concept with care can shape innovation, reduce risks, and lead to meaningful impact in society.
"""

    final_output = f"""
============================
🔹 SHORT SUMMARY: {title.upper()}
============================
{short_bullets}

============================
🔸 MEDIUM SUMMARY: {title}
============================
{medium_section}

============================
🔶 DETAILED SUMMARY: {title}
============================
{detailed_section}
"""
    return final_output
```

```python
sample_transcript = """
Artificial Intelligence is revolutionizing the world of education. From personalized learning experiences to automated grading,
AI technologies are enabling teachers to focus more on student engagement. However, ethical concerns such as algorithmic bias
and data privacy remain key challenges. Governments and institutions must develop policies to ensure AI is used responsibly
while maximizing its benefits in the classroom.
"""

# Run the function
summary_output = generate_3_level_summary(sample_transcript)
print(summary_output)
```

```python
# Save the output to a text file
with open("sample_summary.txt", "w", encoding='utf-8') as file:
    file.write(summary_output)
```

```python
def generate_3_level_summary_with_smart_title(transcript):
    # Step 1: Generate title
    title = generate_title(transcript)

    # Step 2: Generate summaries at 3 levels
    short_raw = summarizer(transcript, max_length=40, min_length=15, do_sample=False)[0]['summary_text']
    medium_raw = summarizer(transcript, max_length=100, min_length=60, do_sample=False)[0]['summary_text']
    long_raw = summarizer(transcript, max_length=200, min_length=100, do_sample=False)[0]['summary_text']

    # Step 3: Format the outputs
    short_bullets = short_raw.replace('. ', '.\n🔸 ')
    short_summary = f"🔸 {short_bullets}"
    medium_summary = f"""
🧩 **Concept Breakdown**
{medium_raw}

💡 **Example**
For instance, imagine applying this concept to a real-life scenario such as...
"""
    detailed_summary = f"""
📘 **Detailed Overview**
{long_raw}

🧠 **What This Means**
This topic connects to broader implications like ethics, usability, or future innovation.

🔚 **Conclusion**
To summarize, this subject plays a crucial role in real-world applications, demanding thoughtful understanding and responsible implementation.
"""

    # Final output
    return f"""
📌 **Short Summary: {title}**
{short_summary}

---

📌 **Medium Summary: {title}**
{medium_summary}

---

📌 **Detailed Summary: {title}**
{detailed_summary}
"""
```

```python
from transformers import pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
title_generator = pipeline("summarization", model="facebook/bart-large-cnn")

def generate_title(transcript):
    result = title_generator(transcript, max_length=15, min_length=5, do_sample=False)
    return result[0]['summary_text'].strip()
```

```python
sample_text = """
Artificial Intelligence (AI) refers to the simulation of human intelligence in machines that are programmed to think like humans and mimic their actions.
It can also be applied to any machine that exhibits traits associated with a human mind such as learning and problem-solving.
The ideal characteristic of artificial intelligence is its ability to rationalize and take actions that have the best chance of achieving a specific goal.
Machine learning, a subset of AI, refers to the concept that computer programs can automatically learn from and adapt to new data without being assisted by humans.
Deep learning techniques enable this automatic learning through the absorption of huge amounts of unstructured data such as text, images, or video.
"""
# Run summarizer function with your Whisper transcript
summary_output = generate_3_level_summary_with_smart_title(sample_text)

# Print to verify in Colab
print(summary_output)

# Save to a text file
with open("summary_output.txt", "w", encoding="utf-8") as f:
    f.write(summary_output)

# Download the file to your local system
from google.colab import files
files.download("summary_output.txt")
```
